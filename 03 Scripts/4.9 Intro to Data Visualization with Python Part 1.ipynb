```python
import pandas as pd
import numpy as np

# Load your dataframes
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_optimized.pkl'))
df_customers = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'customers.csv'))

# Define the chunk size
chunk_size = 1_000_000  # Adjust this number based on your available memory

# Function to split dataframe into chunks
def split_dataframe(df, chunk_size):
    chunks = [df.iloc[i:i + chunk_size] for i in range(0, df.shape[0], chunk_size)]
    return chunks

# Split the large dataframe into chunks
ords_prods_chunks = split_dataframe(ords_prods_merge, chunk_size)
```

```python
# Initialize an empty list to hold the merged chunks
merged_chunks = []

# Merge each chunk with the customer dataframe
for chunk in ords_prods_chunks:
    merged_chunk = chunk.merge(df_customers, on='user_id', how='left', indicator=True)
    merged_chunks.append(merged_chunk)
```

```python
# Concatenate all merged chunks into a single dataframe
ords_prods_all = pd.concat(merged_chunks, ignore_index=True)
```
